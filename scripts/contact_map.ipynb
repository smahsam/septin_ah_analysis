```python
import MDAnalysis as mda
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import MDAnalysis.analysis.distances as distance_array
from MDAnalysis.analysis import distances
```

```python
Z= '1_curved15_45A'
#Z= '2_scratch15_30A'
#Z= '3_bound15_15A'
#Z= '4_tandem_30A'
#Z= '5_tandem_15A'
#Z= '6_tandem_CT2_30A'
#Z= '7_tandem_ex_CT2_30A'
N=5

traj_file = f'{Z}/N{N}/traj_continuous_v1_400_reduced.xtc'
tpr_file  = f'{Z}/N{N}/step7_400.nowat.tpr'

u = mda.Universe(tpr_file, traj_file)
protca = u.select_atoms('name CA')
#print(len(u.trajectory))
protca.atoms.center_of_mass()
```

```python
cutoff=7.5
D = []
for ts in u.trajectory[1500:4000]:    #last 700, 2000:4000
    DIST = distances.distance_array(protca.positions,protca.positions, ts.dimensions)
    D.append(DIST<cutoff)
D = np.array(D)
Dmean = np.mean(D, axis=0)
```
